### 5. Create features and target data, Only select features data that are highly correlated with target data.

```python
X = df.drop("Churn", axis=1)
Y = df.Churn
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_regression

#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)

fit = bestfeatures.fit(X,Y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores
```

```python
X = df[["Total day minutes","Number vmail messages","Total day charge","International plan","Total eve minutes","Customer service calls"]]
Y = df.Churn
```

```python
Y
```

### 6. Scale the target data (churn)

```python
from sklearn.preprocessing import StandardScaler

standardscaler = StandardScaler()
X = standardscaler.fit_transform(X)
X
```

### 7. Check the shape of both training data and testing data

```python
from sklearn.model_selection import train_test_split

np.random.seed(42)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

### 8. Apply Logistic regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=5000)
model.fit(X_train, Y_train)
    
print("Training Score: ",model.score(X_train, Y_train))
print("Testing Score: ",model.score(X_test, Y_test))
```

```python
y_preds = model.predict(X_test)
y_preds[:5]
```

### 9. Display confusion matrix

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_preds,Y_test)
```

```python
sns.heatmap(confusion_matrix(y_preds,Y_test), annot=True, cbar=False, cmap='viridis', fmt='d')
sns.set(font_scale=1.8)
plt.title("Actual v/s Predicted Values")
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
```

### 10. Perform Hyper parameter tuning

```python
# define models and parameters
hyper_logmodel = LogisticRegression()

# Create a hyperparameter grid for LogisticRegression 
param_grid = [
    {'penalty' : ['l1','l2','elasticnet','none'],
     'C' : np.logspace(-4,4,20),
     'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
     'max_iter' : [100,1000,2500,5000]
    }
]
```

```python
hyper_logmodel.get_params().keys()
```

```python
from sklearn.model_selection import GridSearchCV

# define grid search
grid_search = GridSearchCV(estimator = hyper_logmodel,
                           param_grid = param_grid,
                           n_jobs = -1, 
                           cv = 5)

grid_result = grid_search.fit(X_train, Y_train)

#summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

### 12. Check the model score of both training and testing data

```python
print("Training Score: ",grid_search.score(X_train, Y_train))
print("Testing Score: ",grid_search.score(X_test, Y_test))
```
